Theoretical research of deflection coefficient of EM-wave in case of reflection from different structures on water surface:
    

```python
def F_h(tetha, epsC1, eps2, tanTH2):
    G1 = sqrt(epsC1)*sin(tetha)-sqrt(complex(eps2[i],eps2[i]*tanTH2[i])-epsC1*(cos(tetha))**2)
    G2 = sqrt(epsC1)*sin(tetha)+sqrt(complex(eps2[i],eps2[i]*tanTH2[i])-epsC1*(cos(tetha))**2)
    G = G1/G2
    return G;
def F_v(tetha, epsC1, eps2, tanTH2):
    G1 = sqrt(epsC1)*sqrt(complex(eps2[i],eps2[i]*tanTH2[i])-epsC1*(cos(tetha))**2)-complex(eps2[i],eps2[i]*tanTH2[i])*sin(tetha)
    G2 = sqrt(epsC1)*sqrt(complex(eps2[i],eps2[i]*tanTH2[i])-epsC1*(cos(tetha))**2)+complex(eps2[i],eps2[i]*tanTH2[i])*sin(tetha)
    Gv = G1/G2
    return Gv;
```

```python
def F2_h(tetha, lambd, dmax, epsC1, eps2, tanTH2, eps3, tanTH3):
    G1 = sqrt(epsC1)*sin(tetha)-sqrt(complex(eps2[i],eps2[i]*tanTH2[i])-epsC1*(cos(tetha))**2)
    G2 = sqrt(epsC1)*sin(tetha)+sqrt(complex(eps2[i],eps2[i]*tanTH2[i])-epsC1*(cos(tetha))**2)
    G12 = G1/G2
    G21 = sqrt(complex(eps2[i],eps2[i]*tanTH2[i])-(cos(tetha))**2)-sqrt(complex(eps3[i],eps3[i]*tanTH3[i])-((cos(tetha))**2))
    G22 = sqrt(complex(eps2[i],eps2[i]*tanTH2[i])-(cos(tetha))**2)+sqrt(complex(eps3[i],eps3[i]*tanTH3[i])-((cos(tetha))**2))
    G23 = G21/G22
    beta = 2*np.pi*(d/lambd[i])*sqrt(complex(eps2[i],eps2[i]*tanTH2[i])-(cos(tetha))**2)
    Gd1 = (G12+G23*np.exp(2*1j*beta))
    Gd2 = (1+G12*G23*np.exp(2*1j*beta))
    Gd = Gd1/Gd2
    return Gd;
def F2_v(tetha, lambd, dmax, epsC1, eps2, tanTH2, eps3, tanTH3):
    G1v = sqrt(epsC1)*sqrt(complex(eps2[i],eps2[i]*tanTH2[i])-epsC1*(cos(tetha))**2)-complex(eps2[i],eps2[i]*tanTH2[i])*sin(tetha)
    G2v = sqrt(epsC1)*sqrt(complex(eps2[i],eps2[i]*tanTH2[i])-epsC1*(cos(tetha))**2)+complex(eps2[i],eps2[i]*tanTH2[i])*sin(tetha)
    G12v = G1v/G2v
    beta = 2*np.pi*(d/lambd[i])*sqrt(complex(eps2[i],eps2[i]*tanTH2[i])-(cos(tetha))**2)
    G21v = complex(eps2[i],eps2[i]*tanTH2[i])*sqrt(complex(eps3[i],eps3[i]*tanTH3[i])-(cos(tetha))**2)-complex(eps3[i],eps3[i]*tanTH3[i])*sqrt(complex(eps2[i],eps2[i]*tanTH2[i])-(cos(tetha))**2)
    G22v = complex(eps2[i],eps2[i]*tanTH2[i])*sqrt(complex(eps3[i],eps3[i]*tanTH3[i])-(cos(tetha))**2)+complex(eps3[i],eps3[i]*tanTH3[i])*sqrt(complex(eps2[i],eps2[i]*tanTH2[i])-(cos(tetha))**2)
    G23v = G21v/G22v
    Gdv = (G12v+G23v*np.exp(2*1j*beta))/(1+G12v*G23v*np.exp(2*1j*beta))
    return Gdv;
```

```python
def F3_h(tetha, lambd, dmax, epsC1, eps2, tanTH2, eps3, tanTH3):
    G1 = sqrt(epsC1)*sin(tetha)-sqrt(complex(eps2[i],eps2[i]*tanTH2[i])-epsC1*(cos(tetha))**2)
    G2 = sqrt(epsC1)*sin(tetha)+sqrt(complex(eps2[i],eps2[i]*tanTH2[i])-epsC1*(cos(tetha))**2)
    G12 = G1/G2
    G21 = sqrt(complex(eps2[i],eps2[i]*tanTH2[i])-(cos(tetha))**2)-sqrt(complex(eps3[i],eps3[i]*tanTH3[i])-((cos(tetha))**2))
    G22 = sqrt(complex(eps2[i],eps2[i]*tanTH2[i])-(cos(tetha))**2)+sqrt(complex(eps3[i],eps3[i]*tanTH3[i])-((cos(tetha))**2))
    G23 = G21/G22
    beta = 2*np.pi*(dmax/lambd[i])*sqrt(complex(eps2[i],eps2[i]*tanTH2[i])-(cos(tetha))**2)
    Gd1 = (G12+G23*np.exp(2*1j*beta))
    Gd2 = (1+G12*G23*np.exp(2*1j*beta))
    Gd = Gd1/Gd2
    return Gd;
def F3_v(tetha, lambd, dmax, epsC1, eps2, tanTH2, eps3, tanTH3):
    G1v = sqrt(epsC1)*sqrt(complex(eps2[i],eps2[i]*tanTH2[i])-epsC1*(cos(tetha))**2)-complex(eps2[i],eps2[i]*tanTH2[i])*sin(tetha)
    G2v = sqrt(epsC1)*sqrt(complex(eps2[i],eps2[i]*tanTH2[i])-epsC1*(cos(tetha))**2)+complex(eps2[i],eps2[i]*tanTH2[i])*sin(tetha)
    G12v = G1v/G2v
    beta = 2*np.pi*(dmax/lambd[i])*sqrt(complex(eps2[i],eps2[i]*tanTH2[i])-(cos(tetha))**2)
    G21v = complex(eps2[i],eps2[i]*tanTH2[i])*sqrt(complex(eps3[i],eps3[i]*tanTH3[i])-(cos(tetha))**2)-complex(eps3[i],eps3[i]*tanTH3[i])*sqrt(complex(eps2[i],eps2[i]*tanTH2[i])-(cos(tetha))**2)
    G22v = complex(eps2[i],eps2[i]*tanTH2[i])*sqrt(complex(eps3[i],eps3[i]*tanTH3[i])-(cos(tetha))**2)+complex(eps3[i],eps3[i]*tanTH3[i])*sqrt(complex(eps2[i],eps2[i]*tanTH2[i])-(cos(tetha))**2)
    G23v = G21v/G22v
    Gdv = (G12v+G23v*np.exp(2*1j*beta))/(1+G12v*G23v*np.exp(2*1j*beta))
    return Gdv;
```
